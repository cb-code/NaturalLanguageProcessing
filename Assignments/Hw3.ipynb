####  **Use BeautifulSoup or another HTML parsing package to extract text from the article.**
---

```python
def tag_visible(element):
    if element.parent.name in ['style', 'script', 'head', 'title', 'meta', '[document]']:
        return False
    if isinstance(element, Comment):
        return False
    return True

def text_from_html(body):
    soup = BeautifulSoup(body, 'html.parser')
texts = soup.findAll(text=True)
    visible_texts = filter(tag_visible, texts)  
    return u" ".join(t.strip() for t in visible_texts)
```

```python
html = request.urlopen('https://www.cnbc.com/2019/01/17/netflix-price-hike-helps-disney-upcoming-streaming-service-analyst.html').read()
text = text_from_html(html)

print(text)
```

>**1. Use re (regular expression) package to:**

>   * Find all matches of $ amounts in the article

```python
print(re.findall('\$(.+?) ', text))
```

>   * Substitute all numbers with # character and print the output

```python
print(re.sub(r'[0-9]','#', text))
```

>   * Count (using regular expressions) ”Netflix” and “Disney” mentions

```python
print(Counter(re.findall('Disney', text)))
print(Counter(re.findall('Netflix', text)))
```

> **2. Use NTLK and/or Spacy tokenization features to:**

>    * Tokenize sentences and words

```python
# tokenize sentences
sentences = sent_tokenize(text)

# tokenize words
tokens = word_tokenize(text)

print(tokens)
```

>    * Remove all English stop words

```python
# removing all english stop-words

stopwordsList = []
stopwordsList = stopwords.words('english')

nonstop_words = [word for word in tokens if word not in stopwordsList]
print(nonstop_words)
```

>    * List and count n-grams for any given input n

```python
n = 3

# printing list and count of n-grams for n
for item in ngrams(tokens, n):
    print(item)
    
print(Counter(ngrams(tokens, n)))
```

>    * Lemmatize and deduplicate unigrams into a vocabulary of terms.

```python
# lemmatize unigrams into vocab of terms

# function from: 
# https://www.machinelearningplus.com/nlp/lemmatization-examples-python/

from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet

def get_wordnet_pos(word):
    """Map POS tag to first character lemmatize() accepts"""
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}

    return tag_dict.get(tag, wordnet.NOUN)

lemmatizer = WordNetLemmatizer()
vocab_list = []

for word in tokens:
  vocab = lemmatizer.lemmatize(word, get_wordnet_pos(word))
  vocab_list.append(vocab)

# deduplicate unigrams into a vocabulary of terms

vocab_terms = list(set(vocab_list))
print(vocab_terms)
```

>    * Print bigrams and trigrams in the first 5 sentences

```python
POS_tokens = word_tokenize(sentences[5])

# printing bigrams in first 5 sentences
print("Bigrams: ")
for bigram in ngrams(POS_tokens, 2):
    print(bigram)

# printing trigrams in first 5 sentences
print("Trigrams: ")
for trigram in ngrams(POS_tokens, 3):
    print(trigram)
```

>    * Print POS tags in the first 5 sentences

```python
POS_tokens = word_tokenize(sentences[5])
sentence_pos = pos_tag(POS_tokens)

print(sentence_pos)
```

```python
grammar = "NP: {<DT>?<JJ>*<NNP>}"
cp = RegexpParser(grammar)
print(cp.parse(sentence_pos))
```
